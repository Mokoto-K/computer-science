# Exploratory Data Analysis

The first phase of any data science project is exploratory data analysis. This involves getting to know the important features of the data. During this process, be open to accepting unexpected features.
* The data types in columns 
* The granularity of data in rows
* Distributions of quantitative data and measurements such as center and spread.
* Relationships between values and items in the dataset.

## Locating and Downloading Datasets

With the advent of data driven technologies, data science datasets are becoming more accessible, larger, and more specific. In this notebook, we're focusing on two datasets, both obtained from kaggle. You can find a large range of other data science datasets on these website:
* www.kaggle.com/datasets
* azure.microsoft.com/en-au/services/open-datasets
* archive.ics.uci.edu/ml/datasets.php
* en.wikipedia.org/wiki/List_of_datasets_for_machine-learning_research

```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt

def load_data(DATASET_PATH):
    return pd.read_csv(DATASET_PATH)
```

### Dataset 1: World Happiness Report up to 2020
You can access this dataset here: www.kaggle.com/mathurinache/world-happiness-report 
> This is an example of a dataset that is:
* High quality
* Numerical

```python
DATASET_PATH = './datasets/happiness/2015.csv'

#create pandas object
happiness = load_data(DATASET_PATH)
```

```python
#you can change the number in the brackets to increase/decrease how much data is shown
happiness.head()
```

```python
happiness.info()
```

```python
happiness.describe()
```

```python
happiness["Family"].hist()
plt.show()
```

```python
happiness.hist(bins=50, figsize=(20,15))
plt.show()
```

```python
happiness.plot(kind="scatter", x="Family", y="Happiness Score")
plt.show()
```

```python
happiness.plot(kind="scatter", x="Trust (Government Corruption)", y="Happiness Score")
plt.show()
```

```python
attributes = ["Happiness Score", "Economy (GDP per Capita)", "Family",
"Health (Life Expectancy)"]
pd.plotting.scatter_matrix(happiness[attributes], figsize=(12, 8))
plt.show()
```

```python
happiness.plot(kind="scatter", x="Economy (GDP per Capita)", y="Happiness Score",
s=50, c="Generosity", cmap=plt.get_cmap("jet"), colorbar=True)
plt.show()
```

```python
corr_matrix = happiness.corr()
corr_matrix["Happiness Score"].sort_values(ascending=False)
```

### Dataset 2: Police Shootings
You can access this dataset here: www.kaggle.com/washingtonpost/police-shootings
> This is an example of a dataset that is:
* Moderate quality
* Categorical

```python
DATASET_PATH = './datasets/police_shooting/fatal-police-shootings-data.csv'
police = load_data(DATASET_PATH)
```

```python
police.head()
```

```python
police.info()
```

```python
police.describe()
```

**Summary of people who were fleeing/mode of fleeing**

```python
police["flee"].value_counts()
```

**Frequency histogram showing age based distribution of shooting victims**

```python
police["age"].hist()
plt.show()
```

```python
print("Age of youngest victim: ",police["age"].min())
print("Age of oldest victim: ",police["age"].max())
```

**Gender distribution**

```python
police.groupby('gender').size()
```

```python
police['gender'].value_counts(normalize=True) * 100
```

```python
labels = police['gender'].astype('category').cat.categories.tolist()
counts = police['gender'].value_counts()
sizes = [counts[var_cat] for var_cat in labels]
fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%', shadow=False) 
ax1.axis('equal')
plt.show()
```

**Race Distribution**

```python
police['race'].value_counts(normalize=True) * 100
```

**Visualise race distribution where**
 W: White, non-Hispanic B: Black, non-Hispanic A: Asian N: Native American H: Hispanic O: Other None: unknown

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns

race_count = police['race'].value_counts()
sns.set(style="darkgrid")
sns.barplot(race_count.index, race_count.values, alpha=0.9)
plt.title('Frequency Distribution of Races')
plt.ylabel('Number of Occurrences', fontsize=12)
plt.xlabel('Race', fontsize=12)
plt.show()
```
