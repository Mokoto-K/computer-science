## Data Files Demo

In this notebook we will review some examples for importing and exporting data from comma-separated values (CSV) files using the built-in Python functions.

As usual we will start by importing all other Python libraries we need for the fucntions used.

```python
import csv
import math
import matplotlib.pyplot as plt
from sklearn import datasets
```

Now we can proceed to open and read the data into our Python script. The data that we are working with today is a set of GPS datapoints collected from a tracking device attached to a grazing sheep from one of our experiments exploring collective movement in flocking animals.

The data will be stored two arrays `x_set` and `y_set` that will store the x and y coordinates respectively.



```python
x_set = []
y_set = []

with open('sheep.csv', newline='') as f:
    reader = csv.reader(f)
    for row in reader:
        print(row)
        x_set.append(float(row[0]));
        y_set.append(float(row[1]));
```

Now that we have our (quite large) dataset imported to the arrays, we can work with it in our Python program. First we will draw a plot of the data using the `plot( ... )` function.

This is raw data, so there are some errors and missing points that relate to the *quality* of the GPS data (this is influenced number of satellites available, clouds, trees and other sheep). The full dataset has the information that we can use to fix this up. 


```python
plt.figure(figsize=(20,10))        
plt.plot(x_set, y_set, 'r-')
plt.xlabel('x axis')
plt.ylabel('y axis')
```

## How fast does a sheep move?


Now lest go an calculate something interesting from our data. Have you every wondered how fast a sheep moves while walking/grazing? We will investigate this by calcualting the speed that the animal is moving at between each of the data points. 

The example below calcualtes the distance moved by the sheep for each time point, multiplies this by 10, as the data is sampled as 10 samples per second, to get the speed in meters per second. This value is then stored in an array called sheep speed. The `hist( ... )` function is then used to create a histogram to display the distribution of the speeds.  

```python
sheep_speed = []
for t in range(1,len(x_set)):
    d = math.sqrt(((x_set[t-1] - x_set[t]) + (y_set[t-1] - y_set[t]))**2.0)
    # We multiply this value by 10 because the data is sampled at 10hz and we want the output to be m/sec
    d = d*10
    sheep_speed.append(d)

plt.figure(figsize=(20,10))        
plt.hist(sheep_speed, bins=200, range=[0,1])
plt.xlabel('Speed m/sec')
plt.ylabel('Count')
```

## Exporting Data to CSV Files

In this example, we will generate a CSV file that contains the *iris* dataset by exporting it from the data stored within the scikit-learn library. 

First we open the dataset.

```python
# import some data to play with - the iris dataset
iris = datasets.load_iris()
```

Next, we will open a file named *iris.csv* for writing (using the 'w' flag). First we write the header for the file using the `iris.feature_names` list from the dataset. the `writerows( ... )` function is then used to write the dataset to the file.

```python
# Open a connection to the file
with open('iris.csv', 'w', newline='') as f:
    writer = csv.writer(f)
    
    #Write the header row - we get this from the iris.feature_names
    writer.writerow(iris.feature_names)
    
    #Write the data - we get this from iris.data
    writer.writerows(iris.data)


# The data will be stored in iris.csv
```
